#### import modules

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### import csv file format
#### (imported using pandas - good for large data)

```python
data=pd.read_csv('West_Yorkshire_Birds_2015.csv',sep=',')
```

#### Plots

```python
% matplotlib inline
```

#### plot mean values as pie chart

```python
# only plot birds which there is a chance of seeing (<1)

selected_values=data.mean_birds > 1.00
labels=data.species[selected_values]
mean_birds=data.mean_birds[selected_values]

plt.pie(mean_birds,labels=labels)
plt.title("The chance of seeing the most common birds in 2015")
```
