```python
# csv file format
# county, species, mean, rank, % gardens
# imported using pandas - good for large data
```

```python
data=pd.read_csv('West_Yorkshire_Birds_2015.csv',sep=',')
```

```python
print(data.columns)
```

```python
data[data.county=="West Yorkshire"]
```

```python
# to only show data for West Yorkshire
dataset = data[data.county=="West Yorkshire"]

selected_values=dataset.mean_birds > 1.00
labels=dataset.species[selected_values]
mean_birds=dataset.mean_birds[selected_values]

plt.pie(mean_birds,labels=labels)
```
